```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sounddevice as sd

plt.rcParams.update({'font.size': 15})
```

#### Create Chirp Signal

```python
samplesPerSecond = 1000
duration = 3
totalSamples = int(duration * samplesPerSecond)

t = np.arange(totalSamples) / samplesPerSecond
f0 = 50
f1 = 250
t1 = 2

x_chirp = np.cos(2*np.pi*t*(f0+(f1-f0)*np.power(t, 2)/(3*t1**2))) #create Chirp Signal

```

#### Plot Signals

```python

plt.figure(figsize=(17, 4))
# Time Plot
plt.subplot(1, 2, 1)
plt.plot(t, x_chirp)
plt.ylabel("Signal")

# Frequency Plot
plt.subplot(1, 2, 2)
plt.specgram(x_chirp, NFFT=128, Fs=samplesPerSecond, noverlap=120, cmap='jet_r')
plt.colorbar()
plt.ylabel("Frequency (Hz)")
plt.xlabel("Time (s)")

#plt.ylim(0, 1.5*omega2 )
#plt.xlim(min(t), max(t))
```
